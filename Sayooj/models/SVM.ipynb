```python
import pandas as pd
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from sklearn import svm
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
data=pd.read_csv('fake_genuine_merge.csv',header=0)
data.head(10)
```

```python
dropping=pd.read_csv('fake_genuine_merge.csv')
dropping.drop('username',inplace=True,axis=1)
dropping.drop('full_name',inplace=True,axis=1)
data=dropping
data.head()
```

```python
Y=data["is_fake"]
X=data.loc[:, data.columns != 'is_fake']
print(Y.head())
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.40)
x_train
```

```python
svc = svm.SVC()
svc.fit(x_train, y_train)
```

```python
y_pred = svc.predict(x_test)
print(y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
result = accuracy_score(y_test,y_pred)
print("Accuracy:",result*100)
```
