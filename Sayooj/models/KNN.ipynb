```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
data=pd.read_csv('fake_genuine_merge.csv',header=0)
data.head(10)
```

```python
dropping=pd.read_csv('fake_genuine_merge.csv')
dropping.drop('username',inplace=True,axis=1)
dropping.drop('full_name',inplace=True,axis=1)
data=dropping
data.head()
```

```python
Y=data["is_fake"]
X=data.loc[:, data.columns != 'is_fake']
print(Y.head())
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.40)
x_train
```

```python
scaler = StandardScaler()#gaved best accuracy than normalisation
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
x_train
```

```python
classifier = KNeighborsClassifier(n_neighbors = 5)
classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test)
len(y_pred)
```

```python
len(y_test)
```

```python
result = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(result)
print("\n")
result2 = accuracy_score(y_test,y_pred)
print("Accuracy:",result2*100)
```

```python
high=-1
for i in range(1,13):
    classifier = KNeighborsClassifier(n_neighbors = i)
    classifier.fit(x_train, y_train)
    y_pred = classifier.predict(x_test)
    results = accuracy_score(y_test,y_pred)
    if((results)*100 > high):
        high=(results)*100
        output=i
print("best k="+str(output))
```
